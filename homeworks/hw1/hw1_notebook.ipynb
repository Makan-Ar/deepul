## Part (a) Fitting a Histogram

Let $\theta = (\theta_0, \dots, \theta_{d-1}) \in \mathbb{R}^d$ and define the model $p_\theta(x) = \frac{e^{\theta_x}}{\sum_{x'}e^{\theta_{x'}}}$

Fit $p_\theta$ with maximum likelihood via stochastic gradient descent on the training set, using $\theta$ initialized to zero. Use your favorite version of stochastic gradient descent, and optimize your hyperparameters on a validation set of your choice.

**You will provide these deliverables**


1.   Over the course of training, record the average negative log-likelihood (nats / dim) of the training data (per minibatch) and test data (for your entire test set). Code is provided that automatically plots the training curves. 
2.   Report the final test set performance of your final model
3. Plot the model probabilities in a bar graph with $\{0,\dots,d-1\}$ on the x-axis and a real number in $[0,1]$ on the y-axis.




### Solution
Fill out the function below and return the necessary arguments. Feel free to create more cells if need be.

```python
def q1_train(theta, train_data, optimizer):
    theta.train()
    
    train_loss = []
    for x in train_data:
        x = x.to(DEVICE)
        
        optimizer.zero_grad()
        loss = theta.loss(x)
        loss.backward()
        optimizer.step()
        
        train_loss.append(loss.item())
    
    return np.mean(train_loss)
```

```python
def q1_eval(theta, data):
    theta.eval()
    
    losses = []
    with torch.no_grad():
        for x in data:
            x = x.to(DEVICE)
            loss = theta.loss(x)
            losses.append(loss.item())
    
    return np.mean(losses)
```

```python
def q1_trainer(theta, train_data, dev_data, test_data, epochs):
    
    optimizer = optim.Adam(theta.parameters(), lr=0.1, betas=(0.9, 0.999), eps=1e-08, weight_decay=0)
    
    train_loss = []
    dev_loss = []
    test_loss = []
    
    init_test_loss = q1_eval(theta, test_data)
    test_loss.append(init_test_loss)
    
    pbar = tqdm(total=epochs)
    for epoch in range(epochs):
        epoch_train_loss = q1_train(theta, train_data, optimizer)
        epoch_dev_loss = q1_eval(theta, dev_data)
        epoch_test_loss = q1_eval(theta, test_data)
        
        train_loss.append(epoch_train_loss)
        dev_loss.append(dev_loss)
        test_loss.append(epoch_test_loss)
        
        pbar.update()
        pbar.set_description(f'Train loss: {epoch_train_loss:.3e} - Dev: {epoch_dev_loss:.3e}')
    pbar.close()
    
    return train_loss, dev_loss, test_loss
```

```python
def q1_get_data_loader(x, batch_size):
    x = torch.tensor(x, dtype=torch.long)
    return DataLoader(x, batch_size=batch_size, shuffle=True)
```

```python
class Q1Model(nn.Module):
    def __init__(self, d):
        super().__init__()
        self.theta = nn.Parameter(torch.zeros(d), requires_grad=True)
    
    def loss(self, x):
        logits = self.theta.repeat(x.shape[0], 1)
        lossfunc = nn.CrossEntropyLoss(reduction='mean')
        return lossfunc(logits, x)
```

```python
def q1_a(train_data, test_data, d, dset_id):
    """
    train_data: An (n_train,) numpy array of integers in {0, ..., d-1}
    test_data: An (n_test,) numpy array of integers in {0, .., d-1}
    d: The number of possible discrete values for random variable x
    dset_id: An identifying number of which dataset is given (1 or 2). Most likely
             used to set different hyperparameters for different datasets

    Returns
    - a (# of training iterations,) numpy array of train_losses evaluated every minibatch
    - a (# of epochs + 1,) numpy array of test_losses evaluated once at initialization and after each epoch
    - a numpy array of size (d,) of model probabilities
    """
      
    train_data, dev_data = train_test_split(train_data, test_size=0.2, random_state=GLOBAL_SEED)
    train_dloader = q1_get_data_loader(train_data, batch_size=64)
    dev_dloader = q1_get_data_loader(dev_data, batch_size=len(dev_data))
    test_dloader = q1_get_data_loader(test_data, batch_size=len(test_data))
    
    theta = Q1Model(d).to(DEVICE)
    
    train_loss, dev_loss, test_loss = q1_trainer(theta, train_dloader, dev_dloader, test_dloader, epochs=20)
    
    probs = F.softmax(theta.theta, dim=0).cpu().detach().numpy()
    return np.array(test_loss), np.array(test_loss), probs
```

## Part (b) Fitting Discretized Mixture of Logistics

Let us model $p_\theta(x)$ as a **discretized** mixture of 4 logistics such that $p_\theta(x) = \sum_{i=1}^4 \pi_i[\sigma((x+0.5 - \mu_i)/s_i) - \sigma((x-0.5-\mu_i)/s_i)]$

For the edge case of when $x = 0$, we replace $x-0.5$ by $-\infty$, and for $x = d-1$, we replace $x+0.5$ by $\infty$.

You may find the [PixelCNN++](https://arxiv.org/abs/1701.05517) helpful for more information on discretized mixture of logistics.

**Provide the same set of corresponding deliverables as part (a)**

### Solution
Fill out the function below and return the necessary arguments. Feel free to create more cells if need be.

```python
class MixtureOfLogistics(nn.Module):
    def __init__(self, d, n_mixture=4):
        super().__init__()
        
        self._d = d
        self._n_mixture = n_mixture
        
        self._pi = nn.Parameter(torch.rand(self._n_mixture), requires_grad=True)
        self._mu = nn.Parameter(torch.rand(self._n_mixture), requires_grad=True)
        self._s = nn.Parameter(torch.rand(self._n_mixture), requires_grad=True)
    
    def forward(self):
        pi = F.softmax(self._pi, dim=0)
        s = F.relu(self._s)
        return pi, self._mu, s
        
    def loss(self, x):
        x = x.unsqueeze(1)        
        pi, mu, s = self.forward()

        x_plus = x + 0.5
        x_plus[x == self._d - 1] = float('inf')
        sig_plus = F.sigmoid(x_plus - mu) / s
        
        x_minus = x - 0.5
        x_minus[x == 0] = float('-inf')
        sig_minus = F.sigmoid(x_minus - mu) / s
        
        loglikelihood = torch.log(torch.sum(pi * (sig_plus - sig_minus), dim=1))
        loss = -torch.mean(loglikelihood)
        return loss
```

```python
def q1_b(train_data, test_data, d, dset_id):
    """
    train_data: An (n_train,) numpy array of integers in {0, ..., d-1}
    test_data: An (n_test,) numpy array of integers in {0, .., d-1}
    d: The number of possible discrete values for random variable x
    dset_id: An identifying number of which dataset is given (1 or 2). Most likely
           used to set different hyperparameters for different datasets

    Returns
    - a (# of training iterations,) numpy array of train_losses evaluated every minibatch
    - a (# of epochs + 1,) numpy array of test_losses evaluated once at initialization and after each epoch
    - a numpy array of size (d,) of model probabilities
    """
    
    train_data, dev_data = train_test_split(train_data, test_size=0.2, random_state=GLOBAL_SEED)
    train_dloader = q1_get_data_loader(train_data, batch_size=64)
    dev_dloader = q1_get_data_loader(dev_data, batch_size=len(dev_data))
    test_dloader = q1_get_data_loader(test_data, batch_size=len(test_data))
    
    model = MixtureOfLogistics(d).to(DEVICE)
    
    train_loss, dev_loss, test_loss = q1_trainer(model, train_dloader, dev_dloader, test_dloader, epochs=20)
    
#     probs = F.softmax(theta.theta, dim=0).cpu().detach().numpy()
    return np.array(test_loss), np.array(test_loss), 2
```
